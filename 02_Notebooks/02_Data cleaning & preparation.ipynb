## World Bank

In this section, we are dealing with NaNs and performing data cleaning on the dataset obtained from the World Bank.

As we assess the dataset, we've observed that the majority of indicators do not have data available prior to 1981. To address this issue, we have chosen to partition the data into two distinct datasets. The first dataset exclusively focuses on GDP information, specifically from 1961 onwards. The second dataset incorporates both GDP data and other relevant indicators, starting from 1981 and onwards. This strategic division allows us to effectively manage missing data from earlier periods while ensuring that our analysis is based on the most complete and relevant information available.

Since we lack GDP data, we will replace it with the mean and then calculate GDP based on this replacement value.

```python

def growth_calculation(gdp_year_1, gdp_year_2):
    gdp_percentage_growth = ((gdp_year_2 - gdp_year_1)/gdp_year_1)*100
    
    return gdp_percentage_growth

def gdp_calculation(gdp_year_2, growth_year_2):
    gdp_year1 = (gdp_year_2*100)/(100+growth_year_2)
    
    return gdp_year1

mean = wb_gdp['CAN_gdp_per_capita_%_growth'].mean()
wb_gdp['CAN_gdp_per_capita_%_growth'] = wb_gdp['CAN_gdp_per_capita_%_growth'].fillna(mean)
    
for i in reversed(range(36)):
      wb_gdp.loc[i, 'CAN_gdp_per_capita_constant_us$'] = gdp_calculation(wb_gdp['CAN_gdp_per_capita_constant_us$'][i+1],wb_gdp['CAN_gdp_per_capita_%_growth'][i+1])
```

Some subjects are not relevant to predict a recession / global crisis, so we will drop those subjects.

```python
# I will drop the subjects which do not belong to the list of most relevant for global crisis
subject_to_drop = ['Population',
                   'General government revenue',
                   'Gross domestic product corresponding to fiscal year, current prices',
                   'Current account balance',
                   'Gross domestic product, deflator',
                   'Gross domestic product based on purchasing-power-parity (PPP) share of world total',
                   'Implied PPP conversion rate',
                   'Inflation, end of period consumer prices',
                   'General government total expenditure',
                   'General government net lending/borrowing',
                   'General government primary net lending/borrowing',
                   'Total investment',
                   'Gross national savings',
                   'Volume of Imports of goods',
                   'Volume of exports of goods',
                   'General government structural balance']
```

```python
for column in imf.columns:
    for i in range(len(imf)):
        try:
            if imf[column].isna()[i]==True and imf.iloc[i]['Subject Descriptor'] in subject_to_drop:
                imf = imf.drop(index=i, axis=1)
        except:
            pass
imf = imf.reset_index(drop=True)
imf.isna().sum()
```

```python
imf.shape
```

There are still a lot of NaNs. To get rid off those NaNs, we will use the mean values. The reason to use values until 2019 has to do with potential 'noise' after COVID and the estimated values thereafter.

```python
     
years_to_2019 = ['1980', '1981','1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990','1991', '1992',
               '1993', '1994', '1995', '1996', '1997', '1998', '1999','2000', '2001', '2002', '2003', '2004', '2005',
               '2006', '2007', '2008','2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017','2018',
               '2019']

for column in imf.columns:
    for i in range(len(imf)):
        try:
            if imf[column].isna()[i]==True:
                mean = imf.loc[i][years_to_2019].mean(skipna=True)
                imf.loc[i,column] = mean
        except:
            pass

imf = imf.reset_index(drop=True)
imf.isna().sum()
```
